## Impuestos

Los impuestos son cuotas que se deben pagar al gobierno para aportar al gasto público o del estado; estos son obligatorios y deben pagarse en determinada cantidad según la persona o empresa, dependiendo el nivel de ingresos. Estos funcionan para:
- Administración pública
- programas sociales
- infraestructura
(BBVA, 2022).
